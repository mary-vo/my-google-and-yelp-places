```python
import pandas as pd
import json
import requests
```

```python
yelp_df = pd.read_excel("yelp_html.xlsx").head(10)
headers = {"Accept": "*/*", "Authorization": "Bearer "}

yelp_api_response=[]
for url in yelp_df['Business URL']:
    business_id = url.split('/')[-1]
    business_url= f"https://api.yelp.com/v3/businesses/{business_id}"

    r = requests.get(url=business_url, headers=headers)
    response_dict = json.loads(r.text)
    yelp_api_response.append(response_dict)
yelp_api_df = pd.DataFrame(yelp_api_response)
yelp_api_df.to_excel("yelp_response.xlsx")
```

```python
google_csv_df = pd.read_excel('google_csv.xlsx').head(10)

google_csv_api_response=[]

for url in google_csv_df['URL']:
    url_string = 'https://maps.googleapis.com/maps/api/place/details/json?'
    cid = url.split(':')[-1]
    params = {'cid':cid, 'key':''}

    response = requests.get(url = url_string, params=params)
    response_dict = response.json()
    google_csv_api_response.append(response_dict)

normalize = [pd.json_normalize(item) for item in google_csv_api_response]

google_csv_api_df = pd.concat(normalize, ignore_index=True)
google_csv_api_df.to_excel("google_csv_response.xlsx")
```
